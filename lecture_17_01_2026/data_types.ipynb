## 1) Numeric Types

### int (Integer)

```python
a = 10
b = -5
```

```python
print("Integer:", a, b)
```

### float (Floating-Point Number)

```python
x = 10.5
y = -3.14

print("Float:", x, y)
```

### complex

Numbers with real + imaginary parts.

```python
z = 2 + 3j

print("Complex Number:", z)
```

## 2. Boolean Type

### bool

Represents True or False.

```python
is_active = True
is_logged_in = False
print("Boolean:", is_active, is_logged_in)
```

## 3. Text Type

### str

Sequence of characters (string).

```python
msg = 'Hello'
print("String:",  msg)
```

## 4. Sequence Types

These hold multiple values.

### a) list (Mutable)

Ordered, ( Mutable ) changeable, Allows duplicates.

```python
nums = [1, 2, 3]
mixed = [1, "A", 3.5]
nums.append(4)
print("List:", nums, mixed)
```

### b) tuple (Immutable)

Ordered, cannot be changed, allows duplicates.

```python
point = (10, 20)
print("Tuple:", point)
```

### c) range

Represents a sequence of numbers (used in loops).

```python
r = range(1, 5)  # 1,2,3,4
print("Range:", list(r))
```

## 5. Mapping Type

### dict

Key-value pairs.

```python
fruit = {
    "name": "Mango",
    "color": "Yellow"
}
print("Dictionary:", fruit)
```

## 6. Set Types

Used to store unique values.

### a) set

- Unordered, mutable, unique items.

```python
colors = {"red", "green", "green", "blue"}
print("Set:", colors)
```

```python
colors.add("yellow")
print("Set after adding yellow:", colors)
```

### b) frozenset (Immutable set)

- Cannot be changed once created.

```python
f = frozenset([1, 2, 3])
print("Frozen Set:", f)
```

```python
f.add(4)  # This will raise an error because frozenset is immutable
```

## 7. Binary Types
- Used for handling bytes or binary data.

### a) bytes

Immutable sequence of bytes.

```python
b = b"Hello"

print("Bytes:", b)
print(type(b))
```

### b) bytearray

Mutable version of bytes.

```python
ba = bytearray(b"Hello")
print("Bytearray before modification:", ba)
ba[0] = 73  # modify
print("Bytearray after modification:", ba)

print(type(ba))
```

### c) memoryview

Allows access to slices of binary data without copying.

```python
mv = memoryview(b"5")

print("Memory View:", mv.tobytes())
print(id(mv))
print(type(mv))
```

## 8. Special Type

## NoneType

Represents "no value".

```python
x = None
print("NoneType:", x)
```
